# Multi Armed Bandits Simulation
In this assignment you will simulate multi-armed bandits with different behaviours using numpy, there are a total of 3 different behaviours: 

One with probabilistic behaviour, which gives different outputs with a probability \
One which gives you a noisy reward, for each input \
Note : All the rewards (or mean rewards) should be randomly generated and should not be hardcoded

```python
# Import necessary libraries here
import numpy as np
import matplotlib.pyplot as plt
```

The probabilistic behaviour means for each of the arms there will be a randomly generated set of rewards, {r_1, r_2, ..., r_k} and randomly generated set of probabilities {p_1, p_2, ..., p_k}. After pulling any arm, you should choose one of the rewards in its reward set with its corresponding probability and return it. The choice of value of k is upto you, but you will need numpy for k > 2.

```python
class MAB_probabilistic:
    def __init__(self, num_action : int = 10) -> None:
        self.num_action = num_action
        self.rewards_possible = []
        self.probabilities=[]
        self.reward_given = 0 #No reward yet as there is no action
        for i in range(num_action):
            temp = np.random.randint(1,11,size=self.num_action) #rewards from 1 to 10
            self.rewards_possible.append(list(temp)) #For each action(arm) we are defining a random set of rewards and probabilities
            temp = np.random.rand(self.num_action)
            self.probabilities.append(list(temp/np.sum(temp)))
    def act(self, action : int) -> float:
        if self.num_action in range(1,self.num_action+1):
            reward_given = np.random.choice(self.rewards_possible[action-1], p=self.probabilities[action-1])
            #print("Valid Action!",end=' ')
            return reward_given
        else:
            #print("Invalid Action!",end=' ')
            return 0 # reward is 0 for invalid action
```

The noisy policy means for each arm, you should return a reward chosen from an underlying probability distribution (like gaussian, you can use numpy for this purpose) with a randomly generated mean and the given variance (passed to __init__)

```python
class MAB_noisy:
    def __init__(self, noise_variance : float, num_action : int = 10) -> None:
        self.num_action = num_action
        self.noise_variance = noise_variance
        self.means = np.random.randn(num_action)
    def act(self, action : int) -> float:
        if self.num_action in range(1,self.num_action+1):
            required_mean = self.means[action]
            noisy_reward = np.random.normal(loc=required_mean, scale=np.sqrt(self.noise_variance))
            return noisy_reward
        else:
            #print("Invalid Action!",end=' ')
            return 0 # reward is 0 for invalid action
```

Initialize MAB_probabilistic with number of actions 5. Perform an action for a fixed number of time and plot the average of first n rewards against n. Make such graphs for all 5 actions in single plot.

```python
MAB = MAB_probabilistic(5)
#The possible actions are 1,2,3,4,5
#So, the possible rewards in each arm range from 1 to 5 and their probabilities are random

no_of_actions = 1000
average_rewards1 = []
average_rewards2 = []
average_rewards3 = []
average_rewards4 = []
average_rewards5 = []
for i in range(1, no_of_actions + 1):
    average_reward = np.mean([MAB.act(1)] + average_rewards1)  # always performing action 1
    average_rewards1.append(average_reward)
for i in range(1, no_of_actions + 1):
    average_reward = np.mean([MAB.act(2)] + average_rewards2)  # always performing action 2
    average_rewards2.append(average_reward)
for i in range(1, no_of_actions + 1):
    average_reward = np.mean([MAB.act(3)] + average_rewards3)  # always performing action 3
    average_rewards3.append(average_reward)
for i in range(1, no_of_actions + 1):
    average_reward = np.mean([MAB.act(4)] + average_rewards4)  # always performing action 4
    average_rewards4.append(average_reward)
for i in range(1, no_of_actions + 1):
    average_reward = np.mean([MAB.act(5)] + average_rewards5)  # always performing action 5
    average_rewards5.append(average_reward)
import matplotlib.pyplot as plt
plt.plot(range(1, no_of_actions + 1), average_rewards1)
plt.plot(range(1, no_of_actions + 1), average_rewards2)
plt.plot(range(1, no_of_actions + 1), average_rewards3)
plt.plot(range(1, no_of_actions + 1), average_rewards4)
plt.plot(range(1, no_of_actions + 1), average_rewards5)
plt.xlabel('Number of Trials')
plt.ylabel('Average Reward')
plt.title(f'Average Reward vs. Number of Trials for {no_of_actions} actions')
plt.show()
```
